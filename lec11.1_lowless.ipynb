```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi , 2*np.pi , 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x ,y)
```

```python
noise = np.random.uniform(-.2 , .2 , 100)
```

```python
y_mod = y + noise
```

```python
plt.scatter(x , y_mod)
```

```python
import statsmodels.api as sm
```

```python
lowess = sm.nonparametric.lowess
```

```python
gen = lowess(y_mod , x)
```

```python
x_gen , y_gen = gen[: , 0] , gen[: , 1]
```

```python
plt.scatter(x , y_mod)
plt.scatter(x_gen , y_gen)
```
