```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X , Y)
```

#### Goal 
#### given a functin f(x) , we want to find the value of x that minimize f

```python
plt.plot(X,Y)
plt.ylabel("F(x)")
plt.xlabel("x")
plt.show()
```

```python
x = 0 
lr = 0.1
error = []
# 50 steps in the downhill direction 
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x=x-lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```
