# Locally Weighted Regression

### 1) Read and Normalise the Dataset 

### 2) Generate W for every query point

### 3 ) No training is involved , directly make predictions using closed form solution we derived ! 

### 4) find the best value of tau (Bandwidth Parameter ) [Cross validation ]

## Loading Data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dfx = pd.read_csv("/Users/habiburrahmankhan/datascience/datasets/LocallyWeightedRegression/weightedX.csv")
dfy = pd.read_csv("/Users/habiburrahmankhan/datascience/datasets/LocallyWeightedRegression/weightedY.csv")
```

```python
X = dfx.values
Y = dfy.values
print(X.shape)
print(Y.shape)
```

```python
X = (X - X.mean()) / X.std()
```

```python
plt.style.use("seaborn")
plt.title("Normalise Data ")
plt.scatter(X , Y)
#plt.xlim(-2 , 2)
#plt.ylim(-2  ,4)
plt.show()
```

###   find out W (m*m)

```python
def getW(queryPoint , X , tau):
    M = X.shape[0]
    W = np.mat(np.eye(M))
    for i in range(M):
        xi  =X[i]
        x = queryPoint
        W[i , i] = np.exp(np.dot((xi - x) , (xi - x).T) / (-2*tau*tau) ) 
    return W
```

```python
X = np.mat(X)
Y = np.mat(Y)
M  =X.shape[0]
W = getW(-1 ,X ,100)
print(W.shape)
print(W)
```

### Make prediction

```python
def predict(X , Y , query_x , tau):
    ones = np.ones((M,1))
    X_ = np.hstack((X , ones))
    qx  =np.mat([query_x  ,1])
    W = getW(qx , X_, tau)
    #print(X.shape)
    #print(X_[: 5 , : ])
    
    theta = np.linalg.pinv(X_.T*(W*X_))*(X_.T*(W*Y))
    #print(theta.shape)
    pred = np.dot(qx , theta)
    return theta , pred
    
    
```

```python
theta , pred = predict(X , Y, 1.0, 1.0)
```

```python
def plotPred(tau):
    X_test = np.linspace(-2 , 2, 20)
    Y_test = []
    for xq in X_test:
        theta , pred = predict(X , Y ,xq , tau)
        Y_test.append(pred[0][0])
    Y_test = np.array(Y_test)
    Xo = np.array(X)
    Yo = np.array(Y)
    plt.title("Tau / bandwidth %.2f"%tau )
    plt.scatter(Xo , Yo)
    plt.scatter(X_test , Y_test , color="red")
    plt.show()              
```

```python
tau = [.1 , .2 ,.5  ,1 , 5 ,10]
for i in tau:
    plotPred(i)
```
