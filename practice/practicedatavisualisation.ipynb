## Line Plot

```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
y1 = X**2
y2 = 2*X + 3
print(X)
print(y1)
print(y2)
```

```python
plt.plot(X , y1 , color='r' , label="apple" , marker= "*")
plt.plot(X , y2 , color='g' , label = "kiwi"  , marker="o")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruit price over time ")
plt.show()
```

## Scatter Plot 

```python
plt.style.use("seaborn")

plt.scatter(X , y1 , color='r' , label="apple" , marker= "*")
plt.scatter(X , y2 , color='g' , label = "kiwi"  , marker="o")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Fruit price over time ")
plt.legend()
plt.show()
```

## bar graph 

```python
x = np.array([1,2, 3])**2
plt.bar(x-.25 , [10,20 , 15] , width=0.5 , label="Current Year " , tick_label = ["Gold " , "Platinum" , "Silver"])
plt.bar(x+.25 , [10,20 , 15] , width=0.5 , label="Current Year " )
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.xlim(0,10)
plt.ylim(0,40)
plt.legend()
plt.show()
```

## PIE CHART

```python
subject = "Maths" , "Chem" , "Physics" , "English"
weightes = [20,10 , 15 , 5]
plt.pie( weightes , labels = subject , autopct = "%1.1f%%", explode=(0,0,0.1,0))
plt.show()
```

```python
freq_title = {}
for title in titles:
    length = len(title)
    if freq_title.get(length) is None:
        freq_title[length] =1 
    else:
        freq_title[length] +=1
```

```python
print(freq_title)
```

```python
Y = np.array(list(freq_title.values()))
```

## Multivarient normal Distribution : - 

```python
#Bivariate
mean = np.array([0.0 , 0.0])
cov = np.array([[1,0.5 ] , [0.5,1]])
mean2 = np.array([5.0 , 6.0])
cov2 = np.array([[1.3,0.2 ] , [0.2,1.1]])
dist = np.random.multivariate_normal(mean ,cov , 500 )
dist2 = np.random.multivariate_normal(mean2 ,cov2 , 500 )
print(dist.shape)
```

```python
plt.scatter(dist[:,0] , dist[: , 1])
plt.scatter(dist2[:,0] , dist2[: , 1])
plt.show()
```
