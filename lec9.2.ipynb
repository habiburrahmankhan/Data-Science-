```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from PIL import Image
strange = Image.open("drstrange.jpeg")
img = np.array(strange)
```

```python
plt.imshow(img)
```

```python
X = img.reshape((-1,3))
```

```python
X.shape
```

```python
model = KMeans(8)
```

```python
model.fit(X)
```

```python
model.cluster_centers_
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
plt.imshow(pallets.reshape(1,-1,3))
```

```python
gen = np.zeros_like(X , dtype = np.uint8)
```

```python
for i in range(gen.shape[0]):
    gen[i] = pallets[model.labels_[i]]
```

```python
plt.imshow(gen.reshape(img.shape))
```
