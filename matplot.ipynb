```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X =np.linspace(-10 ,10 , 10)
```

```python
y = X**2
```

```python
z = X**3
```

```python
plt.figure()
plt.plot(X,y)

plt.show()
```

```python
plt.figure()
plt.plot(X,y)
plt.plot(X,z)
plt.show()
```

```python
plt.figure()
plt.plot(X,z)
plt.show()
```

```python
plt.figure()
plt.plot(X,y , label= "Square")
plt.plot(X,z,label= "Cube")
plt.xlim(-10 , 10)
plt.ylim(-100 , 100)
plt.xlabel("this is for input")
plt.ylabel("this is for output")
plt.title("poore 20 number ")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.scatter(X,y , label= "Square" , color="red")
plt.scatter(X,z,label= "Cube" , color="yellow")
plt.xlim(-10 , 10)
plt.ylim(-100 , 100)
plt.xlabel("this is for input")
plt.ylabel("this is for output")
plt.title("poore 20 number ")
plt.legend()
plt.show()
```

```python
angs = np.linspace(0 , 2 *np.pi ,1000)
X =5*np.cos(angs)
y = 5*np.sin(angs)
```

```python
plt.figure()
plt.plot(X, y)
plt.xlim(-6 , 6)
plt.ylim(-6 , 6)
plt.title("circle")
plt.xlabel("this is for input")
plt.ylabel("this is for output")
plt.legend()
plt.show()
```

```python
angs = np.linspace(0 , 20*np.pi,1000)
rad = np.linspace(0 , 20,1000)
X =rad*np.cos(angs)
y = rad*np.sin(angs)
```

```python
plt.figure()
plt.plot(X, y)
plt.title("circle")
plt.xlabel("this is for input")
plt.ylabel("this is for output")
plt.legend()
plt.show()
```

```python
X = np.linspace(0,2*np.pi , 1000)
plt.figure()
plt.subplot(2,2,1)
plt.plot(X , np.sin(X))
plt.subplot(2,2,2)
plt.plot(X , np.cos(X))
plt.subplot(2,1,2)
plt.plot(X ,np.tan(X))
plt.show()
```
