```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import make_regression 
```

```python
bias=100
X , y , coef = make_regression(n_features= 1 , noise =4 , random_state = 42 , shuffle=False , coef=True , bias=bias)
```

```python
y.shape
```

```python
plt.scatter(X ,y)
```

### slope means coeff and intercept means bias 

```python
x_min , x_max = X.min() , X.max() 
```

```python
X_gen = np.linspace(x_min , x_max , 10)
```

```python
y_gen = X_gen*coef + bias
```

```python
plt.scatter(X ,y)
plt.plot(X_gen ,y_gen)
```
